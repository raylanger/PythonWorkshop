# Matplotlib Tutorial

```python
# Importing in Jupyter Notebooks
import matplotlib.pyplot as plt

import numpy as np
```

## The Basics

### Plot types

```python
xdata = np.arange(0,5)
ydata = xdata**2

#Line plot is plt.plot
plt.plot(xdata,ydata) #Make the plot
plt.show() #Show the plot

#Scatter plot is plt.scatter
plt.scatter(xdata,ydata)
plt.show()

#Bar plot is plt.bar
plt.bar(xdata,ydata)
plt.show()

#List of more complex plot types can be found here: https://matplotlib.org/stable/plot_types/index.html
```

### Customization

There are an innumerable number of ways to customize a plot, and we will not be able to go over everything. Some key things that people typically want to customize are:
<br> <br>
colour - to customize the colour of the points or lines, use c = 'colour', where colour is some string which indicates the colour you actually want to use. Common colours can be indicated using only one letter (i.e. b for blue, r for red, k for black, etc.), and matplotlib has a list of colours with specific names that can be found here: https://matplotlib.org/stable/gallery/color/named_colors.html. If you want to use a very specific colour, use a # followed by the hexadecimal value for that string, or the RGB values.
<br>
Example:

```python
colorlist = ['k','r','b','aquamarine','hotpink','#8db6e5',(141/256, 182/256, 229/256)]
for i in range(0,7):
    plt.plot(range(5),i*np.ones(5), c = colorlist[i])
plt.show()
```

Note that, because I did not put plt.show() until after the for loop, all the lines were plotted on the same graph.
<br><br>
For a scatter plot, you can make each point be a different colour using an array of colours

```python
plt.scatter(xdata,ydata,c = ['k','r','b','aquamarine','hotpink'])
plt.show()
```

For bar plots or scatter plots, you may want to have the edge colour be different from the fill colour. To specify the edge colour, use ec = 'colour', and for fill color (or face color) use fc = 'colour'.
Example:

```python
plt.bar(xdata,ydata, ec = 'k', fc = 'r')
plt.show()
```

Another useful general customization option is the transparency, which is referred to as alpha. A fully opaque graph has an alpha of 1, whereas a fully transparent one has an alpha of 0.
<br>
Example:

```python
a = np.arange(0,1.1,0.2)
for i in range(6):
    plt.plot(range(5),i*np.ones(5),c = 'k', alpha = a[i])
plt.show()
```

For line plots: <br>
Line thickness - can be adjusted using linewidth or lw = float, which specifies the thickness of the line in pts, with the default being 1.5. <br><br>
Line style - can be adjusted using linestyle or ls = str, where str is one of '-' (solid line), '--' (dashed line), ':' (dotted line), '-.' (alernating dashes and dots), '' or ' ' (no line, for usage with markers).<br><br>
Example:

```python
thicknesses = [0.5,1.5,3,5,10]
for i in range(5):
    plt.plot(range(5),i*np.ones(5),c = 'k',lw = thicknesses[i])
plt.show()

styles = ['-','--',':','-.','']
for i in range(5):
    plt.plot(range(5),i*np.ones(5),c = 'k',ls = styles[i])
plt.show()
```

For scatter plots: <br>
marker size - set marker size using s = float, where float represents the area of the point in pt^2.
<br><br>
marker shape - set marker shape using marker = str, where str is some string representing the shape of the marker. Possible options include 'o' (default, circle), '*' (star), 's' (square), '+' (plus), and 'x' (xs). 
<br><br>
Note: Markers can be added to line plots by using marker = str as well. Then, marker size can be adjusted using ms, and marker edge and face colour can be adjusted through mec and mfc. 
<br><br>
Example:

```python
sizes = [2,5,10,20,50]
for i in range(5):
    plt.scatter(range(5),i*np.ones(5),c = 'k',s = sizes[i])
plt.show()

markers = ['o','*','s','+','x']
for i in range(5):
    plt.scatter(range(5),i*np.ones(5),c = 'k',marker = markers[i])
plt.show()

plt.plot(xdata,ydata, marker = 'o')
plt.show()
```

### Other things you need to know

#### Texts
title - title can be added through plt.title(str)<br>
labels - labels can be added using plt.xlabel(str) or plt.ylabel(str)<br>
general text - text can be added anywhere through plt.text(x,y,str). x and y represent the coordinates of the text, str represents the message. <br>
font size can be adjusted through fontsize = float in all cases <br><br>

Example:


```python
plt.plot(xdata,ydata)
plt.title(r'$y = x^2$')
plt.xlabel(r'$x$')
plt.ylabel(r'$y$')
plt.text(2,8,r'$y = x^2$')
plt.show()

fig = plt.figure()
ax = fig.add_subplot()
ax.plot(xdata,ydata)
ax.text(0,1,'a',transform = ax.transAxes)
#Note: r'$$' used to allow us to write text using LaTeX notation
```

#### Axes
limits - xlimit set through plt.xlim(left,right), ylimit set through plt.ylim(bottom,top)<br>
ticks - xticks set through plt.xticks(locations,labels), yticks set through plt.yticks(locations,labels)<br>

Example:

```python
plt.plot(xdata,ydata)
plt.xlim(0,4)
plt.xticks([0,1,2,3,4],['a','b','c','d','e'])
plt.ylim(0,16)
plt.yticks([0,1,4,9,16],['cat','dog','mouse','bear','lion'])
plt.show()
```

### Adding a legend

legend - a legend can be added through plt.legend(handles,labels). A legend can be added in many different ways, and neither handles or labels is always necessary to specify. In general, handles refers to the lines that you want to label, and labels refers to the names you want to give them. 

Examples:

```python
colorlist = ['black','red','blue']

#Method 1 - label specified in plt.plot
plt.plot(range(5),np.ones(5), c = colorlist[0], label = colorlist[0])
plt.plot(range(5),2*np.ones(5), c = colorlist[1], label = colorlist[1])
plt.plot(range(5),3*np.ones(5), c = colorlist[2], label = colorlist[2])
plt.legend()
plt.show()

#Method 2 - label specified in plt.legend
plt.plot(range(5),np.ones(5), c = colorlist[0])
plt.plot(range(5),2*np.ones(5), c = colorlist[1])
plt.plot(range(5),3*np.ones(5), c = colorlist[2])
plt.legend(labels = [colorlist[0],colorlist[1],colorlist[2]])
plt.show()

#Method 3 - handles used to specify which lines in legend
line1, = plt.plot(range(5),np.ones(5), c = colorlist[0], label = colorlist[0])
line2, = plt.plot(range(5),2*np.ones(5), c = colorlist[1], label = colorlist[1])
plt.plot(range(5),3*np.ones(5), c = colorlist[2], label = colorlist[2])
plt.legend(handles = [line1,line2])
plt.show()

#Method 4 - handles used to specify which lines to label, labels used to specify new names
line1, = plt.plot(range(5),np.ones(5), c = colorlist[0], label = colorlist[0])
line2, = plt.plot(range(5),2*np.ones(5), c = colorlist[1], label = colorlist[1])
plt.plot(range(5),3*np.ones(5), c = colorlist[2], label = colorlist[2])
plt.legend(handles = [line1,line2], labels = ['Lfdsaukfhkdsa', 'fdlkishakf 2'])
plt.show()
```

## Advanced Matplotlib

Very frequently, we will want to have multiple plots arranged together in a certain way. How do we do this?

#### Figures, GridSpecs and Axes

Matplotlib uses a hierarchical approach to fixing this question. At the top of the hierarchy is the figure. The figure holds everything that will be plotted. The next level down is the GridSpec. The Gridspec divides the figure up into cells into which axes will be placed. Generally, this grid is rectangular and uniform, but we will show how to customize it to achieve a desired result. The final part of the hierarchy is the axes. This is the object to which data is actually added. <br>

So what is happening when we call plt.plot()? When we dont specify a figure or axis, matplotlib will either add the plot to the most recent open figure and axis, or if there is no open figure/axis, it will create a new figure and axis to plot onto. This is why running several plt.plot() functions in a row without a plt.show() allows us to plot everything on the same graph. Until we run plt.show(), the figure and axis stay open and continue to be the most recent open figure. Once we run plt.show(), the figure closes, and the next plot will create a new figure and axis.<br>

Now, lets get into how we can manipulate figures, gridspecs, and axes. We will start with a simple case, that is, 2 rows of 2 subplots of equal size. <br>

In general, to create a figure with a uniform gridspec that has a subplot in each cell, we use: <br>
fig, axs = plt.subplots(nrows,ncols)<br>
Note that this function returns both the figure object, which I labelled fig, and *an array* containing each axes in the figure, which I labelled axs. To obtain the axes object in the ith row and jth column, I would use ax = axs[i,j]. <br>

For our problem, we thus want to run plt.subplots(2,2). Lets also assume that we want the subplots to be a bit flatter, so we should adjust the figure size as well, which we can do through plt.subplots(2,2,figsize = [12,8]), which sets figure length to be 12 inches, and figure height to be 8 inches. Since the figure is longer than it is wide, the plots will be longer than they are wide as well.

Example:

```python
fig, axs = plt.subplots(2,2,figsize = [12,8])
colorlist = ['k','r','b','g']
markerlist = ['o','s','+','x']
lwlist = [1,2,3,4]
for e,ax in enumerate(axs.flat):
    ax.plot(xdata,ydata, c = colorlist[e],marker = markerlist[e], lw = lwlist[e])
    #Note how we need to use set_title instead of just title
    ax.set_title('Plot %d' % (e+1))
    #can just use set(), then specify everything you want to set at once
    ax.set(xlabel = 'x', ylabel = 'y', xlim = [0,4], ylim = [0,16])
    
fig.suptitle('2 by 2 subplots')
plt.show()
```

```python
#To prevent overlap, we can usually just use fig.tight_layout() at the end

fig, axs = plt.subplots(2,2,figsize = [12,8])
colorlist = ['k','r','b','g']
markerlist = ['o','s','+','x']
lwlist = [1,2,3,4]
for e,ax in enumerate(axs.flat):
    ax.plot(xdata,ydata, c = colorlist[e],marker = markerlist[e], lw = lwlist[e])
    #Note how we need to use set_title instead of just title
    ax.set_title('Plot %d' % (e+1))
    #can just use set(), then specify everything you want to set at once
    ax.set(xlabel = 'x', ylabel = 'y', xlim = [0,4], ylim = [0,16])
    
fig.suptitle('2 by 2 subplots')
fig.tight_layout()
plt.show()
```

In order to do this, we will need to specify an exact gridspec. This requires us to import the package matplotlib.gridspec, and then we can use the function matplotlib.gridspec.GridSpec(nrows, ncols, width_ratios, height_ratios). The access to width_ratios and height_ratios are what make using the gridspec directly necessary here. I will first give an example of using GridSpec and then discuss how it works.

```python
import matplotlib.gridspec as gs
fig = plt.figure()

grid = gs.GridSpec(3,2,width_ratios = [3,2],height_ratios = [3,2,4.5])

#Add subplot to 1st row, all columns
ax1 = fig.add_subplot(grid[0,:])

#Add subplot spanning 2nd and 3rd row, 1st column
ax2 = fig.add_subplot(grid[1:,0])

#Add subplot in 2nd row, 2nd column
ax3 = fig.add_subplot(grid[1,1])

#Add subplot in 3rd row, 2nd column
ax4 = fig.add_subplot(grid[2,1])

fig.tight_layout()
ax1.plot(xdata,ydata)
plt.show()
```

## Pandas

Pandas is extremely useful for data management. I think of it as like having Excel in your Python workspace. To use pandas, we obviously have to import it first, which we do through <br>
import pandas as pd

The next step after that is to load up our data. Pandas can read data from tons of different formats. The most common formats you will see are python dictionaries, comma-separated values (CSVs), and text values (usually full-width files or FWFs)

```python
import pandas as pd

#Reading data from a python dictionary
hitterstats = {'AVG': [.342,.305,.298,.344],
              'OBP': [.474,.374,.444,.481],
              'SLG': [.690,.555,.607,.634],
              'HR': [714,755,761,521]}
df = pd.DataFrame(hitterstats)
display(df)

df = pd.DataFrame(hitterstats, index = ['Babe Ruth', 'Hank Aaron','Barry Bonds','Ted Williams'])
display(df)

#Reading data from a CSV
df = pd.read_csv('./CompSims/Project2/BBEData.csv', sep = ',')
display(df)

#Reading data from a textfile
#Using read_csv and a different sep
df = pd.read_csv('./NRCan/MagneticResearch/Carisma Data/FCHU/2001/20010301FCHU.MAG', sep = '\s+', header = 39, names = ['DATE','X','Y','Z','F','.']).drop('.', axis = 1)
display(df)

#Using read_fwf
df = pd.read_fwf('./NRCan/MagneticResearch/Carisma Data/FCHU/2001/20010301FCHU.MAG', header = 39, names = ['DATE','X','Y','Z','F'], widths = [14,10,10,10,2])
display(df)
```

```python
#print(df) works, but looks like this:
print(df)
#I prefer to use display
```

The object used by pandas is known as a DataFrame. Each column in a DataFrame is a Series object. We can access each series in a DataFrame through df['column name'] or df.col_name. Note that to use the second option, the column name must have no spaces or dots or other invalid characters.

If you happen to want the values in a column as a list and not a series, then you can use df['column'].values.

Example:

```python
display(df) #Dataframe

display(df.DATE,df['DATE']) #Series associated with 'DATE' column

print(df.DATE.values)
```

You can create a new column through df['new column'] = array, though array must be the same length as the DataFrame. We can also create new columns by manipulating existing columns, i.e. df['new column'] = df['column 1'] + df['column 2'].

Examples:

```python
df['Square Roots'] = np.sqrt(range(len(df)))
df['X+Y'] = df['X'] + df['Y']
display(df)
```

We can isolate certain portions of the dataframe using df.loc[] or df.iloc[]. df.iloc[a:b] will isolate all the rows between the a-th row and the b-th row, including a but not including b (like when you isolate certain parts of an array). If we use df.loc[a:b] instead, this will isolate all rows between the a-th **index** and the b-th **index**. The difference is illustrated below.

```python
testdf = df.loc[10:20] #Could also use iloc here as for df, index matches up with row number
display(testdf)
display(testdf.loc[:5]) #Since the index of testdf starts at 10, this will return an empty dataframe
display(testdf.iloc[:5]) #iloc wil return indices 10 to 14, since these are the first 5 rows of testdf
```

We may also use df.loc[condition] to return all the indices of the df which satisfy that condition. Some examples are illustrated below.

```python
display(df.loc[df['X'] > 8400]) #Basic condition
display(df.loc[df.Y.between(65,66)]) #Between two numbers (could also use two consecutive conditions)
display(df.loc[np.sqrt(df.Z) > 244.57]) #Applying numpy function to a number
display(df.loc[df.DATE > 20010301210000].loc[(df.X + df.Z) < 68150]) #Two consecutive conditons
```

Lets also quickly take a look at what each condition looks like. This is not necessary per se, but better understanding of how this works may help you debug if you get stuck.

```python
np.sqrt(df.X) > 8400
```

Note that, when using 'and' or 'or' in Python, these can only be used to compare the actual values True or False. Thus, you cannot do something like df.loc[df.X>8400 and df.Y.between(65,66)] because df.X>8400 does not return True or False, it returns a *Series* containing True and False. Instead, you would have to use the *bitwise* and/or symbols, i.e. & or |. So, df.loc[df.X>8400 & df.Y.between(65,66)] would work.

You can combine .loc and .iloc with column names to isolate the column as well as the rows which satisfy the condition. With .loc, the column name is used, and with .iloc, the column index is used.

Examples:

```python
display(df.loc[df.X > 8400,'Y'])
display(df.iloc[40:50,3])
```

To access a singular cell, we use df.at[index, column]. Column can be either a number representing the index of the desired column, or a string, representing the name of the desired column.

Examples:

```python
df.at[4000,'X']
```

Values can be set on the isolated parts of the dataframe. For example, there are some errant values in this dataset, which have been marked by the value 99999.99. If we want to set all of these to NaN, such as if we wanted to calculate the mean, we could do that through:

```python
print(np.nanmax(df.X))
df.loc[df.X > 90000,'X'] = np.nan
print(np.nanmax(df.X))

print(df.X.mean())
```

Note that, to obtain the mean of X, I can simply run df.X.mean(). Other statistics can be obtained similarly. To see an overview of the stats for each column of a DataFrame, you can use df.describe().

Example:

```python
df.describe()
```

To sort data, you can just use df.sort_values('column name', inplace = True). inplace = True just says to replace the current df with the sorted df. If this is missing, the df will not be updated. This is useful if you just want to see the sorted df.

Examples:

```python
display(df.sort_values('X'))
display(df)

df.sort_values('X', inplace = True)
display(df)
```

Note that now the indices are out of order. If we want to reset it, we can use df.reset_index(inplace = True,drop = True). inplace does the same thing it did before, and drop just tells the function to get rid of the old indices as well.

Example:

```python
df.reset_index(inplace = True, drop = True)
df
```

If you have a function which only applies to singular values, you can apply it to each value in a column through df.column.apply(fcn), where fcn is the name of the function.

Example:

```python
def goodZ(z):
    if np.isnan(z):
        return 'bad'
    else:
        return 'good'
```

```python
goodZ(df.Z)
```

```python
df.Z.apply(goodZ)
```

To get rid of columns or rows, we use df.drop(index name(s), axis = 0, inplace = True) for rows and df.drop(column name(s), axis = 1). inplace again does the same thing. And the axis parameter is 0 for rows and 1 for columns. 

Examples:

```python
#Dropping a row
display(df)
df.drop(0,axis = 0, inplace = True)
display(df)

#Dropping a column
df.drop('F', axis = 1, inplace = True)
display(df)
```

Finally, if you want to export your data to a file, this can be done easily through df.to_csv(filename). Like with read_csv, you can specify the separator using sep = 'str'.

Example:

```python
df.to_csv('Testdf1.csv', sep = ':')
```
